````python
import pandas as pd
from bs4 import BeautifulSoup
import requests
# just in case if you want to remove string from number
import re # regular expression
````

````python
url='https://www.bmw.co.th/th/all-models.html'
r=requests.get(url)
r.text
````

````python
s=BeautifulSoup(r.text)
s
````

````python
el=s.select("div.cmp-modelcard")
len(el)
````

````python
# el
e=el[0]
print(e.prettify)
````

````python
records=[]
for e in el:
    try:
        model=e.select_one("h5.cmp-modelcard__name").text.strip()
        fuelx=e.select_one("div.cmp-modelcard__fuel-type").text.strip()
        fuel=re.sub(r'•','/',fuelx)
        pricex=e.select_one("span.cmp-modelcard__price").text.strip()
        price=int(re.sub(r'\D','',pricex))
        # print(model,fuel,price)
        records.append([model,fuel,price])
    except:
        continue
records
````

````python
df=pd.DataFrame(records,columns=['Model','Fuel','Price'])
df
````
